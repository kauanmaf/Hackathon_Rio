```python
import folium

# Criando o mapa da cidade do Rio de Janeiro
mapa_rio = folium.Map(location=[-22.9068, -43.1729], zoom_start=11)

def cor_por_gravidade(gravidade):
    if gravidade == 'Baixo':
        return 'blue'
    elif gravidade in ["Medio", "Normal", "Média"]:
        return 'orange'
    elif gravidade == 'Alto':
        return 'red'
    elif gravidade == 'Critico':
        return 'darkred'
    elif gravidade == "Sem_classificacao":
        return 'gray'
    else:
        return 'green'

def cor_por_classe(classe):
    if classe == 'Ponto critico':
        return 'red'
    elif classe == 'Ponto observado':
        return 'orange'
    elif classe == 'Ponto monitorado':
        return 'green'


# Adicionando pontos de dados como marcadores no mapa
for index, row in cor_ocorrencias_pop_filtrado.dropna().iterrows():
    popup = f"{row['pop_titulo']} /// Início: {row['data_inicio']} /// Fim:{row['data_fim']}"
    folium.CircleMarker(location=[row['latitude'], row['longitude']],
                        radius=5,
                        color=cor_por_gravidade(row['gravidade']),
                        fill=True,
                        fill_color=cor_por_gravidade(row['gravidade']),
                        fill_opacity=0.6,
                        popup=popup).add_to(mapa_rio)

for index, row in ra_psalagamento[["classe", "latitude", "longitude", "medida_cor", "causa_alagamento"]].dropna().iterrows():
    popup = f"Causa Alagamento: {row['causa_alagamento']} /// Medida: {row['medida_cor']}"
    folium.CircleMarker(location=[row['latitude'], row['longitude']],
                        radius=5,
                        color=cor_por_classe(row['classe']),
                        fill=True,
                        fill_color=cor_por_classe(row['classe']),
                        fill_opacity=0.6,
                        popup=popup).add_to(mapa_rio)

                    

# Exibindo o mapa com os marcadores
mapa_rio.save('mapa_rio_scatter.html')  # Salva o mapa como um arquivo HTML
```
